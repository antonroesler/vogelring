```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/out/main.csv", index_col="ID")
df.head()
```

```python
df_places = pd.read_csv("../../data/out/orte.csv", delimiter=";", index_col="ID")
df_places.head()
```

```python
df = df.merge(df_places[["Gebiet", "lat", "lon"]], left_on="Ort", right_on="Gebiet", how="left")
```

```python
df.head()
```

```python
# Count occurrences of each RingString and get top 20
top_birds = df["RingString"].value_counts().head(20)
print("Top 20 most frequently observed birds:")
print("\nRank  Ring Number    Species         Observations")
print("-" * 50)
for i, (ring, count) in enumerate(top_birds.items(), 1):
    species = df[df["RingString"] == ring]["Art"].iloc[0]
    print(f"{i:2d}.   {ring:12s}  {species:15s}  {count:4d}")
```

## Analyze for Friends

```python
ring_of_interest = "E09301"


print(f"Found {len(df[df["RingString"] == ring_of_interest])} sightings")
```

```python
# Get the date/location pairs from the filtered dataframe
date_loc_pairs = df[df["RingString"] == ring_of_interest][["Datum", "Ort"]].drop_duplicates()

# Find all rows that match any of these date/location combinations
matching_rows = pd.merge(df, date_loc_pairs, on=["Datum", "Ort"])
matching_rows

ring_counts = matching_rows[matching_rows["RingString"] != ring_of_interest]["RingString"].value_counts()
ring_counts = ring_counts[ring_counts > 5]
ring_counts
```
