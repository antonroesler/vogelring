```python
import pandas as pd
from pathlib import Path
```

```python
root = Path("../../data")
file = root / "main.csv"
```

```python
Path().absolute()
```

```python
df = pd.read_csv(file, sep=";")
df = df.drop(
    [
        "Koordinaten",
        "IngLong",
        "IngLat",
        "Spalte1",
        "Tag",
        "Monat",
        "Jahr",
        "strSpecies",
        "strRingScheme",
        "Nummer",
        "Bein",
        "Ort3",
        "Unnamed: 26",
        "Datum2",
        "dtmDate",
        "strSex",
        "strAge",
        "Alter",
    ],
    axis=1,
)
df.head()
```

```python
def to_int_or_nan(x):
    try:
        return int(x)
    except (ValueError, TypeError):
        return pd.NA


df["Gruppengröße"] = df["Gruppengröße"].apply(to_int_or_nan)

# Show the results
df.head()
```

```python
# Remove quotes from both columns
df["Bemerkung"] = df["Bemerkung"].str.replace('"', "").str.replace("'", "")
df["Melder"] = df["Melder"].str.replace('"', "").str.replace("'", "")
df["Ort"] = df["Ort"].str.replace('"', "").str.replace("'", "")
```

```python
# Drop rows where any of these columns are empty
print(f"Shape before dropping empty rows: {df.shape}")
df = df.dropna(subset=["Art", "RingString", "Ablesung"])

# Show the results and the new shape
print(f"Shape after dropping empty rows: {df.shape}")
```

```python
df.to_csv(root / "out" / "main.csv")
```
